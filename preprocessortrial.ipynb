```python
import os 
import shutil
```

```python
dataset_folder = "/Users/ahmed_ali/Documents/GitHub/Project_DL_UJM/lavender-dataset" 
```

```python
# count the number of images in each folder 
image_counts = {}
new_folder_path = "/Users/ahmed_ali/Documents/GitHub/Project_DL_UJM/lavender_dataset_splitted_stress"
for root, dirs, files in os.walk(dataset_folder):
    for file in files:
        parent_folder = os.path.basename(root)
        print(f"Processing file: {file} in folder: {parent_folder}")
        if file.endswith(".jpg") or file.endswith(".png"):
            # check the file name and split based on it 
            stress_level = file.split("_")[1]
            if stress_level == "S10" and (parent_folder == "T2" or parent_folder == "T4"):
                # save to a new folder 
                # create a folder for S10 if it doesn't exist
                stress_folder = os.path.join(new_folder_path, "stressed")
                if not os.path.exists(stress_folder):
                    os.makedirs(stress_folder)
                    #copy the file to the new folder
                src_path = os.path.join(root, file)
                # folder + file 
                new_file_name = f"{parent_folder}_{file}"
                dst_path = os.path.join(stress_folder, new_file_name)
                shutil.copy(src_path, dst_path)

            if stress_level == "S25" and (parent_folder == "T2" or parent_folder == "T4"):
                # save to a new folder 
                # create a folder for S25 if it doesn't exist
                stress_folder = os.path.join(new_folder_path, "stressed")
                if not os.path.exists(stress_folder):
                    os.makedirs(stress_folder)
                    #copy the file to the new folder
                src_path = os.path.join(root, file)
                new_file_name = f"{parent_folder}_{file}"
                dst_path = os.path.join(stress_folder, new_file_name)
                shutil.copy(src_path, dst_path)

            if (stress_level == "S50" and (parent_folder == "T2" or parent_folder == "T4")) or ((stress_level == "S10" or stress_level == "S25") and (parent_folder == "T0")):
                # save to a new folder 
                # create a folder for S50 if it doesn't exist
                stress_folder = os.path.join(new_folder_path, "not_stressed")
                if not os.path.exists(stress_folder):
                    os.makedirs(stress_folder)
                    #copy the file to the new folder
                src_path = os.path.join(root, file)
                new_file_name = f"{parent_folder}_{file}"
                dst_path = os.path.join(stress_folder, new_file_name)
                shutil.copy(src_path, dst_path)
            
```

```python
# count the number of images in each folder 
image_counts = {}
new_folder_path = "/Users/ahmed_ali/Documents/GitHub/Project_DL_UJM/lavender_dataset_splitted_stress"
for root, dirs, files in os.walk(new_folder_path):
    for file in files:
        if file.endswith(".jpg") or file.endswith(".png"):
            folder = os.path.basename(root)
            if folder not in image_counts:
                image_counts[folder] = 0
            image_counts[folder] += 1
print(image_counts)
```

```python
# loop over the whole dataset and crop them
import cv2
dataset_folder = "/Users/ahmed_ali/Documents/GitHub/Project_DL_UJM/lavender_dataset_splitted_stress"
for root, dirs, files in os.walk(dataset_folder):
    for file in files:
        sample_image_path = os.path.join(root, file)
        image = cv2.imread(sample_image_path)
        height, width, _ = image.shape
        crop_size = 1024

        start_x = (width - crop_size) // 2
        start_y = (height - crop_size) // 2
        cropped_image = image[start_y:start_y+crop_size, start_x:start_x+crop_size]
        # save and replace it the folder
        cv2.imwrite(sample_image_path, cropped_image)
```

```python
# check the maximum and minimum intensity of the images in the dataset
import numpy as np
dataset_folder = "/Users/ahmed_ali/Documents/GitHub/Project_DL_UJM/lavender_dataset_splitted_stress"
max_intensity = 50
min_intensity = 200
for root, dirs, files in os.walk(dataset_folder):
    for file in files:
        if file.endswith(".jpg") or file.endswith(".png"):
            sample_image_path = os.path.join(root, file)
            image = cv2.imread(sample_image_path)
            max_intensity = max(max_intensity, np.max(image))
            min_intensity = min(min_intensity, np.min(image))
print(f"Max intensity: {max_intensity}, Min intensity: {min_intensity}")
```
