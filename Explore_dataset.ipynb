```python
import os 
import cv2
import matplotlib.pyplot as plt
```

```python
dataset_folder = "/Users/ahmed_ali/Documents/GitHub/Project_DL_UJM/lavender-dataset" 
```

```python
# count the number of images in each folder and count how many S50, S10, S25 in each folder 
image_counts = {}
for root, dirs, files in os.walk(dataset_folder):
    for file in files:
        if file.endswith(".jpg") or file.endswith(".png"):
            folder = os.path.basename(root)
            stress_level = file.split("_")[1] 
            name = folder + "_" + stress_level
            if name not in image_counts:
                
                image_counts[name] = 0 
                

            image_counts[name] += 1   

print(image_counts)
            
```

```python
# Image Processor 
# First Crop the image
# Check for the image size 
# loop over all images see if they are the same size if not resize them to the same size
for root, dirs, files in os.walk(dataset_folder):
    for file in files:
        if file.endswith(".jpg") or file.endswith(".png"):
            image_path = os.path.join(root, file)
            image = cv2.imread(image_path)
            if image.shape != (1400, 1400, 3):
                print(f"Image {file} has a different size: {image.shape}")
            # print(image.shape)

```

```python
# sample image 
sample_image_path = "/Users/ahmed_ali/Documents/GitHub/Project_DL_UJM/lavender-dataset/T2/tel11_S25_T4.png"
# crop from the center 1000*1000
image = cv2.imread(sample_image_path)
height, width, _ = image.shape
crop_size = 1000
start_x = (width - crop_size) // 2
start_y = (height - crop_size) // 2
cropped_image = image[start_y:start_y+crop_size, start_x:start_x+crop_size]
print(cropped_image.shape)
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
